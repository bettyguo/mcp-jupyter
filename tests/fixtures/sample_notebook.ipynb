# Customer analysis

Sample notebook used by the integration tests and the demo script.

```python
import pandas as pd
import numpy as np
```

```python
rng = np.random.default_rng(42)
n = 10000
customer_df = pd.DataFrame({
    'id': np.arange(n),
    'age': rng.integers(18, 80, n),
    'tenure_months': rng.integers(0, 120, n),
    'mrr': rng.gamma(2.0, 30.0, n).round(2),
    'country': rng.choice(['US', 'UK', 'DE', 'FR', 'JP'], n),
    'churned': rng.random(n) < 0.12,
})
```
